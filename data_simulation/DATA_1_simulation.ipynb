```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import norm
from datetime import datetime
```

```python
def call_prices(r_list, sigma_list, T_list, N, s0_list, K_list):
    #df = pd.DataFrame(columns=['r','sigma','T','s0','k','t','asset','call'])
    lista = []
    counter = 0
    
    for r in r_list:
        for sigma in sigma_list:
            for T in T_list:
                for s0 in s0_list:
                    for K in K_list:
                        # Generate asset prices
                        asset = [s0*np.exp(sigma*np.sqrt(T*i/N)*np.random.normal(0, 1, size=1)[0] \
                       +(r-sigma**2/2)*T*i/N) for i in range(N)]
                            
                        # Compute call price
                        call = []
                        for i in range(len(asset)):
                            d1= (np.log(asset[i]/K)+(r+(0.5)*(sigma**2))*(T-i*(T/N))) / (sigma*np.sqrt(T-i*(T/N)))
                            d2= (np.log(asset[i]/K)+(r+(-1)*(0.5)*(sigma**2))*(T-i*(T/N))) / \
                            (sigma*np.sqrt(T-i*(T/N)))
                            c = asset[i]*(norm(0,1).cdf(d1))-K*np.exp(-1*r*(T-i*(T/N)))*(norm(0, 1).cdf(d2))
                            #df.loc[counter] = [r,sigma,T,s0,K,i,asset[i],c]
                            lista.append(np.array((r,sigma,T,s0,K,i,asset[i],c)))
                            counter+=1
                            if (counter % 10000)==0:
                                print(counter)
    return lista
```

```python
r_list_1m = np.arange(0.05,1.05,0.1)
sigma_list_1m = np.arange(0.05,1.05,0.1)
T_list_1m = np.arange(0.1,1.1,0.1)
s0_list_1m = np.arange(10,20,1)
K_list_1m = np.arange(10,20,1)
N_1m=10
par = len(r_list_1m)*len(sigma_list_1m)*len(T_list_1m)*len(s0_list_1m)*len(K_list_1m)*N_1m
par
```

```python
start = datetime.now()
data = call_prices(r_list=r_list_1m, sigma_list=sigma_list_1m,\
                   T_list=T_list_1m, N=N_1m, s0_list=s0_list_1m, K_list=K_list_1m)
end = datetime.now()
total_time = (end-start).total_seconds()
```

```python
df = pd.DataFrame(data, columns=['r','sigma','T','s0','k','t','asset','call'])
df
```

```python
df['asset_greater_call'] = [True if df['asset'][i] >= df['call'][i] else False for i in range(df.shape[0])]
df
```
